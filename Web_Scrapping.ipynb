```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.bolsasymercados.es/bme-exchange/es/Negociar/Participantes/Miembros-Renta-Variable"
content = requests.get(url)
soup = BeautifulSoup(content.text, 'html.parser')
# Verifico les taules i les seves classes
tables = soup.find_all('table')
tables
```

```python
#busco les files 
table_rows = soup.find_all('tr', role='row')
table_rows
```

```python
#poso la informació en una taula
data = [] #Crea una llista buida anomenada data on s'emmagatzemaran les dades extretes de la taula. 
for row in table_rows:
    row_data = [] #Crea una llista buida anomenada row_data per emmagatzemar les dades de la fila actual.
    cells = row.find_all('td') #Això retorna una llista de totes les cel·les de la fila.
    for cell in cells:#Itera per cada cel·la en la llista
        row_data.append(cell.get_text(strip=True))
        # obtinc el text dins de la cel·la actual, eliminant qualsevol espai en blanc al principi o al final del text.

    data.append(row_data)#Afegeix la llista row_data a la llista data
headers = ['Nombre', 'Código', 'Dirección Social','Bolsa']
data = data[1:]  # Elimino la primera fila que es el seu (table header)
df = pd.DataFrame(data, columns=headers)
df
```

Amb aquest codi, navega a una URL específica utilitzant el controlador obert. En aquest cas, obre la URL en incognit i crea un objecte al navegador amb webdriver que he instal.lat.
El codi utilitza Selenium per cercar l'element <tbody> amb l'atribut role establert com a "rowgroup" mitjançant el selector CSS tbody[role="rowgroup"]. Això permet localitzar la taula on es troben les dades que volem recopilar.
Faig scroll fins al final per recopilar totes les dades visibles, i després busco elements específics amb selecció CSS per extreure les dades de la taula i crear un DataFrame amb elles. Això permet realitzar web scraping en pàgines web dinàmiques que es generen amb JavaScript i recuperar les dades que volem utilitzar amb Pandas.



```python
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
import time

options = Options() # Posem les Chrome options
options.add_argument("--incognito") #Obrim la pestanya en incognit per fer web scrapping
options.binary_location = "/Applications/Google Chrome.app/Contents/MacOS/Google Chrome"  #Posem on esta installat el nostre google Chrom
driver = webdriver.Chrome(options=options)# Creem un objecte a WebDriver 
driver.get("https://www.bolsasymercados.es/bme-exchange/es/Mercados-y-Cotizaciones/Acciones/Mercado-Continuo/Precios/mercado-continuo")
driver.maximize_window()
time.sleep(1) #posem un retras d'un segon per assegurar-nos que la pàgina es carrega sencera abans de fer el script

# FemScroll fins el final de la pàgina 
previous_height = 0
while True:
    scroll_height = driver.execute_script("return document.documentElement.scrollHeight")
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(1)
    if scroll_height == previous_height: # Si el scroll height no canvia, hem arribat al final de la pàgina
        print('End of page')
        break
    previous_height = scroll_height

# Busquem el codi HTML del body 
body = driver.execute_script("return document.body")
source = body.get_attribute('innerHTML')

# Data scrapping fent cerca l'element <tbody> amb l'atribut role establert com a "rowgroup" utilitzant el selector CSS.
data = []
table = driver.find_element(By.CSS_SELECTOR, 'tbody[role="rowgroup"]')
rows = table.find_elements(By.TAG_NAME, 'tr')

for row in rows:
    company = row.find_element(By.TAG_NAME, 'a').text

    # mirem si a la taula les dades a la fila existeixen abans d'accedir als tags de 'td'
    try:
        row_elements = row.find_elements(By.TAG_NAME, 'td')
        if len(row_elements) > 1:
            price = row_elements[1].text
        else:
            price = ""

        if len(row_elements) > 2:
            change = row_elements[2].text
        else:
            change = ""

        if len(row_elements) > 7:
            date = row_elements[7].text
        else:
            date = ""

        if len(row_elements) > 6:
            value = row_elements[6].text
        else:
            value = ""
        data.append([company, price, change, value])
    except Exception as e:
        print("Error occurred while scraping:", e)

driver.close() # Tanquem el driver

# Create a DataFrame from the data list
df = pd.DataFrame(data, columns=['Company', 'Price', 'Change', 'Value'])
df
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)

try:
    driver.get("https://www.seg-social.es/wps/portal/wss/internet/Trabajadores/PrestacionesPensionesTrabajadores/10963/28393/28396/28475")
    driver.maximize_window()

    # Espera fins que apareguin els elements amb classe "mark_content"
    wait = WebDriverWait(driver, 10)
    elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'mark_content')))

    articles_text = []

    for element in elements:
        # Extreu el codi HTML dels elements amb classe "mark_content"
        html_content = element.get_attribute('outerHTML')

        # Parseja el codi HTML amb BeautifulSoup per eliminar les etiquetes <p></p> i obtenir el text pur
        soup = BeautifulSoup(html_content, 'html.parser')
        text_without_paragraphs = soup.get_text()

        # Afegeix el text sense les etiquetes <p></p> a la llista d'articles
        articles_text.append(text_without_paragraphs)

finally:
    driver.quit()

# Imprimeix el text de tots els articles
for index, article_text in enumerate(articles_text, 1):
    print(f"Article {index}:")
    print(f"\033[1m{article_text}\033[0m")  # Aplica negreta a través de '\033[1m' i torna a l'estil normal amb '\033[0m'
    print("\n")
```
